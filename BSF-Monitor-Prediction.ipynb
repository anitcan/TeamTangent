## Prediction of border intrusion in unmanned areas using satelite images


```python
raster = plt.imread('/Data/spatial/elevation.tif')

plt.imshow(raster)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import rasterio
```

```python
with rasterio.open('/Data/spatial/elevation.tif') as src:
    raster = src.read(1, window=((450, 650), (1200, 1400)))
```

```python
plt.imshow(raster, cmap='gray')
```

```python
import numpy as np
X = np.reshape(raster, (-1, 1))
```

```python
from sklearn.cluster import KMeans
satellite_model = KMeans(n_clusters=8)
```

```python
labels = satellite_model.fit_predict(X)
```
